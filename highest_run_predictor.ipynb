# Task 1

We are predicting runs scored by batsman in every match.<br>
For this we scraped the data of batsman's performance in each match  from espn website.<br>

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Removed unneccasry rows from data.

```python
df = pd.read_csv('/content/Batsman_matches4.csv')
df = df[df['runs'].isin(['absent', 'sub']) == False]
df=df[df['inns'].isin(['Ireland','Bangladesh'])==False]
df.replace('-', 0, inplace=True)
df
```

We have used some features <br>avg_6,<br>avg_4,<br>avg_bf,<br>avg_sr,<br>avg_mins.

The above features description:-<br>
avg_6- It is the average number of sixes hit by batsman per match<br>
avg_4- It is the average number of fours hit by batsman per match<br>
avg_bf- It is the average number of balls faced  by batsman per match<br>
avg_sr- It is the average number of strike rate of  batsman per match which is calculated by <br>
total runs hit  by batsman /total balls faced till now<br>
avg_mins- It is the average mins duration played by batsman per match

Removed unneccasry columns from data.

```python
df=df.drop(['SR','dismissal','odi_id','4s','6s','BF','mins'],axis=1)
df
```

```python
df.dtypes
```

Converted columns into int.

```python
df['runs'] = pd.to_numeric(df['runs']).astype('int')
df['pos'] = pd.to_numeric(df['pos']).astype('int')
df['inns'] = pd.to_numeric(df['inns']).astype('int')
```

```python
df.dtypes
```

Selecting the data that have teams which are playing in world cup 2023

```python
values_to_keep = ['India','Sri Lanka','Bangladesh','Pakistan','England','New Zealand','South Africa','Australia','Netherlands','Afghanistan']
df = df[df['opposition'].isin(values_to_keep)]
```

```python
df.shape
```

Doing the one-hot encoding for dealing with categorical features that important for predicting runs which are:-<br>
player-name of the player<br>
ground-venue of the match<br>
oppsition-opposition team<br>

```python
# One-hot encoding using pandas
one_hot_encoded = pd.get_dummies(df['player'], prefix='player')

# Concatenate the one-hot encoded columns to the original DataFrame
df = pd.concat([df, one_hot_encoded], axis=1)
```

```python
# One-hot encoding using pandas
one_hot_encoded = pd.get_dummies(df['ground'], prefix='ground')

# Concatenate the one-hot encoded columns to the original DataFrame
df = pd.concat([df, one_hot_encoded], axis=1)
```

```python
# One-hot encoding using pandas
one_hot_encoded = pd.get_dummies(df['opposition'], prefix='opposition')

# Concatenate the one-hot encoded columns to the original DataFrame
df = pd.concat([df, one_hot_encoded], axis=1)
```

Extarcting year, month and day from date feature

```python
df['date'] = pd.to_datetime(df['date'], format='%d-%b-%y')

df['timestamp'] = df['date'].astype(int) // 10**9  # Convert to Unix timestamp

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
df
```

We have used randomforestRegressor,lightgbm,XGBregressor,DNN(Deep Neural Networks) models for predicting runs for each batsman in a match and we have got below r2 score for each model

XGBoost- 0.741698412183227<br>
RandomforestRegressor-0.7067971768993471<br>
Lightgbm,XGBregressor-0.5208976825267001<br>
DNN(Deep Neural Networks)-0.8217465896115936<br>

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
import lightgbm as lgb
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score


# Split the data into training and testing sets
# Create a condition for the split
condition_train =(df['year'] >= 2018)
condition_test = (df['year'] == 2023) & (df['month'] >= 10)  # Inverse of the training condition

# Split the data based on the conditions
X_train, X_test, y_train, y_test = (
    df.loc[condition_train],
    df.loc[condition_test],
    df.loc[condition_train],
    df.loc[condition_test]
)
X_test_player=X_test
# X = df[feature_columns]
X_train=X_train.drop(['timestamp','opposition','date','player','ground','inns','pos','runs','year','month','day'],axis=1)
X_test=X_test.drop(['timestamp','opposition','date','player','ground','inns','pos','runs','year','month','day'],axis=1)
y_train=y_train['runs']
y_test=y_test['runs']

scaler =  StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

# XGBoost Regressor
xgb_model = XGBRegressor()
xgb_model.fit(X_train, y_train)
xgb_predictions = xgb_model.predict(X_test)

#Randomforest Regressor
Randomforest_model=RandomForestRegressor(n_estimators=100,max_depth=20)
Randomforest_model.fit(X_train, y_train)
Randomforest_predictions = Randomforest_model.predict(X_test)

# LightGBM Regressor
lgb_model = lgb.LGBMRegressor()
lgb_model.fit(X_train, y_train)
lgb_predictions = lgb_model.predict(X_test)

# DNN using TensorFlow and Keras
model = keras.Sequential([
    keras.layers.Dense(150, activation='relu', input_shape=(X_train.shape[1],)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(1)  # No activation function for regression
])

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])

# Train your DNN model
model.fit(X_train, y_train, epochs=100, batch_size=900)

# Predict with the trained DNN model
dnn_predictions = model.predict(X_test).flatten()

# Evaluate models
print("XGBoost Model:")
print("Mean Squared Error:", mean_squared_error(y_test, xgb_predictions))
print("R-squared Score:", r2_score(y_test, xgb_predictions))

print("\nRandomforest Model:")
print("Mean Squared Error:", mean_squared_error(y_test, Randomforest_predictions))
print("R-squared Score:", r2_score(y_test, Randomforest_predictions))

print("\nLightGBM Model:")
print("Mean Squared Error:", mean_squared_error(y_test, lgb_predictions))
print("R-squared Score:", r2_score(y_test, lgb_predictions))

print("\nDNN Model:")
print("Mean Squared Error:", mean_squared_error(y_test, dnn_predictions))
print("R-squared Score:", r2_score(y_test, dnn_predictions))
```

```python
# Create a new dataframe for predictions
output_df = X_test_player.copy()

# Add the actual target values to the new dataframe
output_df['actual_runs'] = y_test.values

# Add the predicted values from each model
output_df['xgb_predictions'] = xgb_predictions
output_df['Randomforest_predictions'] = Randomforest_predictions
output_df['lgb_predictions'] = lgb_predictions
output_df['dnn_predictions'] = dnn_predictions
```

```python
output_df
```

```python
predicted_runs=output_df[['dnn_predictions','player']]
```

Summing up the total runs scored by each player of the tournament

```python
# Sum the predicted runs for each player
summed_runs = predicted_runs.groupby('player')['dnn_predictions'].sum()

# Sort the dataframe based on the summed values in descending order
sorted_summed_runs = summed_runs.sort_values(ascending=False)

sorted_summed_runs
```
